```python
import pandas as pd
import requests
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Set up API key and endpoint with country parameter
api_key = '99867e3280fc41fcaf88d7160f3e387d'
country = 'in'
url = f'https://newsapi.org/v2/top-headlines?country={country}&apiKey={api_key}&pageSize=100'

# Make request and get response
response = requests.get(url)
data = response.json()
headlines = [article['title'] for article in data['articles']]

# Vectorize headlines
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(headlines)

# Perform K-means clustering
num_clusters = 3
km = KMeans(n_clusters=num_clusters, random_state=0)
km.fit(X)

# Visualize clusters in 2D
labels = km.labels_
centers = km.cluster_centers_

# Get the first two principal components of the vectorized headlines
from sklearn.decomposition import TruncatedSVD
pca = TruncatedSVD(n_components=2)
X_pca = pca.fit_transform(X.toarray())

# Plot the clusters in 2D
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], marker='x', s=200, linewidths=3, color='r')
plt.title('K-means Clustering of News Headlines')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
import pandas as pd
import requests
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
import string
import gensim
from gensim.corpora import Dictionary
from gensim.models import LdaModel
from pprint import pprint
import nltk
# Set up API key and endpoint with country parameter
api_key = '99867e3280fc41fcaf88d7160f3e387d'
country = 'in'
url = f'https://newsapi.org/v2/top-headlines?country={country}&apiKey={api_key}&pageSize=100'

# Make request and get response
response = requests.get(url)
data = response.json()

# Extract headlines
headlines = [article['title'] for article in data['articles']]
nltk.download('stopwords')
nltk.download('wordnet')
# Preprocess headlines
stop_words = set(stopwords.words('english'))
stop_words.add("india")
stop_words.add("indian")
stop_words.add("hindustan")
stop_words.add("ndtv")
stop_words.add("news18")
stop_words.add("time")
stop_words.add("express")
exclude = set(string.punctuation)
lemma = WordNetLemmatizer()

def clean(headline):
    stop_free = " ".join([word for word in headline.lower().split() if word not in stop_words])
    punc_free = "".join(ch for ch in stop_free if ch not in exclude)
    normalized = " ".join(lemma.lemmatize(word) for word in punc_free.split())
    return normalized

preprocessed_headlines = [clean(headline) for headline in headlines]

# Vectorize preprocessed headlines
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(preprocessed_headlines)

# Convert vectorized headlines to gensim dictionary and corpus
tokens = [headline.split() for headline in preprocessed_headlines]
dictionary = Dictionary(tokens)
corpus = [dictionary.doc2bow(token) for token in tokens]

# Perform LDA topic modeling
num_topics = 4
lda = LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics, passes=10)

# Print topics
pprint(lda.print_topics())

# Get the topic distribution for each headline
doc_topics = [lda.get_document_topics(doc, minimum_probability=0.0) for doc in corpus]

# Add topic distribution to dataframe
df = pd.DataFrame({'Headline': headlines})
for i in range(num_topics):
    topic_dist = [0] * len(headlines)
    for j in range(len(headlines)):
        topic_dist[j] = dict(doc_topics[j]).get(i, 0)
    df[f'Topic {i+1}'] = topic_dist

# Display dataframe
print(df)
```

```python
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis

# Visualize topics
vis_data = gensimvis.prepare(lda, corpus, dictionary)
pyLDAvis.display(vis_data)
```

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from scipy.cluster.hierarchy import dendrogram, linkage

# Preprocess the data

preprocessed_headlines_set=set(preprocessed_headlines)
# Vectorize the preprocessed data using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(preprocessed_headlines_set)

# Calculate the pairwise cosine similarity between the documents
similarity_matrix = cosine_similarity(X)

# Perform hierarchical clustering
Z = linkage(similarity_matrix, method='ward')


# Plot the dendrogram
plt.figure(figsize=(20, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
dendrogram(
    Z,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
    truncate_mode='lastp',  # show only the last p merged clusters
    p=37,  # show only the last p merged clusters
    show_contracted=True,  # show the counts on intermediate nodes
)
plt.show()
```
